```python
import os
import numpy as np
import time 

EMBEDDINGS_PATH = '/media/drive2/vkrhk/embeddings/genes'

for embedding_filename in os.listdir(EMBEDDINGS_PATH):
    if embedding_filename.endswith('.npy'):
        path = os.path.join(EMBEDDINGS_PATH, embedding_filename)
        embedding = np.load(path)
        ti_m = os.path.getctime(path)
        m_ti = time.ctime(ti_m)
        print(f'Processing {embedding_filename}: shape: {embedding.shape}, NaN count: {np.sum(np.isnan(embedding))}, Overall count: {embedding.shape[0] * embedding.shape[1]}, {m_ti}')
```
